```python
def mobile_recharge_system():
    """
    Validates a mobile recharge request based on specific plans and minimum amount.
    Allows the user to retry if the plan is invalid.
    """
    valid_plans = [199, 299, 399, 599]

    while True:
        try:
            amount = int(input("Enter recharge amount (â‚¹): "))

            # Rule 1: Recharge amount must be >= 50
            if amount < 50:
                print("âŒ Invalid Amount. Minimum recharge is â‚¹50.")
            
            # Rule 2: Check against valid plans if amount is substantial
            elif amount in valid_plans:
                print(f"âœ… Recharge of â‚¹{amount} Successful!")
                break  # Exit loop on success
            
            else:
                print(f"âš ï¸ Plan not found. Available plans: {valid_plans}")
        
        except ValueError:
            print("âŒ Invalid input. Please enter a numeric value.")
        
        # Retry logic
        retry = input("Do you want to try again? (yes/no): ").strip().lower()
        if retry != 'yes':
            print("Exiting Recharge System.")
            break

# Run the function
mobile_recharge_system()
```

```python
def check_inventory(products):
    """
    Iterates through a product dictionary and prints alerts for low stock.
    Threshold: Stock < 15
    """
    print(f"{'Product':<15} | {'Stock':<10} | {'Status'}")
    print("-" * 40)
    
    for item, quantity in products.items():
        if quantity < 15:
            status = "âš ï¸ Reorder Alert"
        else:
            status = "âœ… Stock OK"
        
        print(f"{item:<15} | {quantity:<10} | {status}")

# Test Data
inventory_data = {
    "Laptop": 10,
    "Mouse": 50,
    "Keyboard": 12,
    "Monitor": 25
}

# Run the function
check_inventory(inventory_data)
```

```python
def process_result(marks_list):
    """
    Calculates average marks from a list.
    Returns: 'Pass' if average >= 50, else 'Fail'.
    """
    if not marks_list:
        return "No marks provided."

    total_marks = sum(marks_list)
    average = total_marks / len(marks_list)
    
    print(f"Total Marks: {total_marks}")
    print(f"Average: {average:.2f}")

    if average >= 50:
        return "Pass"
    else:
        return "Fail"

# Test Data
student_marks = [45, 60, 55, 70, 48]

# Run the function
result = process_result(student_marks)
print(f"Final Result: {result}")
```

```python
def cab_fare_estimator():
    """
    Calculates cab fare:
    - Base fare: â‚¹50
    - Per km: â‚¹12
    - Peak hour: 25% surcharge on total
    """
    BASE_FARE = 50
    PER_KM_RATE = 12

    while True:
        try:
            distance = float(input("\nEnter distance in km: "))
            is_peak = input("Is it peak hour? (yes/no): ").strip().lower()

            # Calculate Standard Fare
            total_fare = BASE_FARE + (distance * PER_KM_RATE)

            # Apply Peak Hour Surcharge
            if is_peak == 'yes':
                total_fare += total_fare * 0.25  # Adding 25%

            print(f"ðŸš– Total Fare: â‚¹{total_fare:.2f}")

        except ValueError:
            print("âŒ Invalid distance entered.")

        # Retry logic
        retry = input("Calculate another fare? (yes/no): ").strip().lower()
        if retry != 'yes':
            print("Thank you for using the Cab Estimator.")
            break

# Run the function
cab_fare_estimator()
```

```python
def check_password_strength(password):
    """
    Checks password strength rules:
    - Length >= 8
    - Contains at least one digit
    - Contains at least one special char (@, #, $)
    """
    special_chars = "@#$"
    has_digit = False
    has_special = False
    
    # Check Length
    if len(password) < 8:
        return "Weak: Password must be at least 8 characters long."
    
    # Loop over string to check content
    for char in password:
        if char.isdigit():
            has_digit = True
        if char in special_chars:
            has_special = True
            
    if has_digit and has_special:
        return "Strong: Password accepted."
    elif not has_digit:
        return "Weak: Must contain at least one digit."
    elif not has_special:
        return "Weak: Must contain at least one special character (@#$)."
    else:
        return "Weak."

# Test Cases
print(check_password_strength("mypassword"))       # Weak (No digit/special)
print(check_password_strength("mypass123"))        # Weak (No special)
print(check_password_strength("MyPass@123"))       # Strong
```

```python
def check_eligibility(attendance_history):
    """
    Calculates attendance percentage.
    Returns: 'Eligible' if % >= 75, else 'Not Eligible'.
    """
    total_days = len(attendance_history)
    if total_days == 0:
        return "No attendance data."

    # Count 'P' (Present) days
    present_days = 0
    for status in attendance_history:
        if status.upper() == 'P':
            present_days += 1
            
    percentage = (present_days / total_days) * 100
    
    print(f"Days Present: {present_days}/{total_days}")
    print(f"Attendance Percentage: {percentage:.2f}%")
    
    if percentage >= 75:
        return "Eligible"
    else:
        return "Not Eligible"

# Test Data
employee_attendance = ["P", "P", "A", "P", "P", "P", "A", "P", "P", "P"]

# Run the function
status = check_eligibility(employee_attendance)
print(f"Exam Eligibility: {status}")
```
