### 1. User Login Check

```python
username = "admin"
password = "1234"

if username == "admin" and password == "1234":
    print("Login Successful")
else:
    print("Invalid Credentials")
```

### 2. Pass / Fail Analyzer

```python
marks = [45, 78, 90, 33, 60]
pass_count = 0
fail_count = 0

for mark in marks:
    if mark >= 50:
        pass_count += 1
    else:
        fail_count += 1

print(f"Total Pass Students: {pass_count}")
print(f"Total Fail Students: {fail_count}")
```

### 3. Simple Data Cleaner

```python
names = [" Alice ", "bob", " CHARLIE "]

cleaned_names = [name.strip().lower() for name in names]
print("Cleaned Names:", cleaned_names)
```

### 4. Message Length Analyzer

```python
messages = ["Hi", "Welcome to the platform", "OK"]

for msg in messages:
    msg_length = len(msg)
    if msg_length > 10:
        print(f"Message: '{msg}' | Length: {msg_length} -> [FLAGGED: Exceeds 10 characters]")
    else:
        print(f"Message: '{msg}' | Length: {msg_length}")
```

### 5. Error Message Detector

```python
logs = ["INFO", "ERROR", "WARNING", "ERROR"]

error_count = logs.count("ERROR")
print(f"Total ERROR entries detected: {error_count}")
```
