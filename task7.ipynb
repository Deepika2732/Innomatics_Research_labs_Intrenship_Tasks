# Smart Light Controller

```python
class SmartLight:
    def __init__(self, name):
        self.name = name
        self.status = "OFF"

    def set_status(self, action):
        # Normalizing input to uppercase for consistency
        self.status = action.upper()

    def display_status(self):
        print(f"{self.name} is {self.status}")

# Execution
light_name = "Bedroom Light"
action = "ON"

my_light = SmartLight(light_name)
my_light.set_status(action)
my_light.display_status()
```

# Employee ID Card System

```python
class Employee:
    def __init__(self, name, emp_id, department):
        self.name = name
        self.emp_id = emp_id
        self.department = department

    def display_card(self):
        print("Employee ID Card")
        print(f"Name: {self.name}")
        print(f"ID: {self.emp_id}")
        print(f"Department: {self.department}")

# Execution
emp = Employee("Rahul", "EMP102", "IT")
emp.display_card()
```

# Mobile Contact Record

```python
class Contact:
    def __init__(self, name, phone):
        self.name = name
        self.phone = phone

    def show_contact(self):
        print("Contact Saved")
        print(f"Name: {self.name}")
        print(f"Phone: {self.phone}")

# Execution
contact1 = Contact("Anita", "9876543210")
contact1.show_contact()
```

# Product Price Tag Generator

```python
class Product:
    def __init__(self, name, price):
        self.name = name
        self.price = price

    def print_tag(self):
        print(f"Product: {self.name}")
        print(f"Price: ₹{self.price}")

# Execution
prod = Product("Headphones", 2499)
prod.print_tag()
```

# Movie Rating Display System

```python
class Movie:
    def __init__(self, name, rating):
        self.name = name
        self.rating = rating

    def display_details(self):
        print(f"Movie: {self.name}")
        print(f"Rating: {self.rating} / 5")

# Execution
movie1 = Movie("Inception", 4.8)
movie1.display_details()
```
