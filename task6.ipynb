```python
def manage_parking(capacity, logs):
    current_parked = 0
    peak_usage = 0

    for log in logs:
        if log == "IN":
            current_parked += 1
        elif log == "OUT":
            current_parked -= 1
        
        # Track the highest number of cars seen at once
        if current_parked > peak_usage:
            peak_usage = current_parked

    status = "Available" if current_parked < capacity else "Full"
    if current_parked > capacity:
        status = "Alert: Over Capacity"

    print(f"Currently Parked Vehicles: {current_parked}")
    print(f"Parking Status: {status}")

# Sample Run
manage_parking(50, ["IN", "IN", "IN", "OUT", "IN", "IN", "OUT"])
```

# Online Food Delivery Time Estimator

```python
def estimate_delivery(distance_km, traffic, weather):
    # Base time: 5 minutes per km
    eta = distance_km * 5 
    
    # Traffic Adjustments
    traffic_delays = {"Low": 0, "Medium": 10, "High": 20}
    eta += traffic_delays.get(traffic, 0)
    
    # Weather Adjustments
    weather_delays = {"Clear": 0, "Rainy": 15, "Stormy": 30}
    eta += weather_delays.get(weather, 0)
    
    print(f"Estimated Delivery Time: {eta} minutes")

# Sample Run
estimate_delivery(8, "High", "Rainy")
```

# Movie Theatre Seat Occupancy Analyzer

```python
def analyze_occupancy(total_seats, booked_seats_list):
    booked_count = len(booked_seats_list)
    occupancy_pct = (booked_count / total_seats) * 100
    
    if occupancy_pct == 100:
        status = "Housefull"
    elif occupancy_pct >= 70:
        status = "Almost Full (Suggest opening additional show)"
    else:
        status = "Seats Available"
        
    print(f"Occupancy: {occupancy_pct}%")
    print(f"Show Status: {status}")

# Sample Run (150 booked seats out of 200)
analyze_occupancy(200, [1] * 150)
```

# Cloud Server Load Classification System

```python
def classify_server_load(readings):
    average_load = sum(readings) / len(readings)
    
    if average_load < 50:
        status = "Normal"
    elif 50 <= average_load <= 80:
        status = "Warning"
    else:
        status = "Critical"
        
    print(f"Average CPU Load: {average_load}%")
    print(f"Server Status: {status}")

# Sample Run
classify_server_load([45, 60, 70, 85, 90])
```

# Smart Classroom Resource Usage Monitor

```python
def monitor_resources(usage_dict):
    threshold = 8
    overused = [res for res, hours in usage_dict.items() if hours > threshold]
    
    alert = "Yes" if overused else "No"
    
    print(f"Overused Resources: {', '.join(overused) if overused else 'None'}")
    print(f"Energy Alert: {alert}")

# Sample Run
monitor_resources({"Projector": 6, "AC": 9, "Lights": 4})
```
