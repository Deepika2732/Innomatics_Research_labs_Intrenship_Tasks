```python
# Input: Taking the sentence as a string
sentence = "python is easy and python is powerful"

# Step 1: Split the sentence into individual words using split()
words_list = sentence.split()

# Step 2: Convert the list of words into a set to remove duplicates
unique_words = set(words_list)

# Step 3: Print the count and the unique words
print(f"Unique words count: {len(unique_words)}")
print(f"Unique words: {unique_words}")
```

```python
# Input: Dictionary of employees and their salaries
employees = {
    "Ravi": 75000,
    "Anita": 68000,
    "Kiran": 72000
}

# Logic: Find the employee with the maximum salary
# We use max() with a key argument to find the key corresponding to the highest value
highest_earning_employee = max(employees, key=employees.get)
highest_salary = employees[highest_earning_employee]

# Output
print(f"Highest Salary: {highest_earning_employee} - {highest_salary}")
```

```python
# Input: List of integers
numbers = [45, 22, 89, 10, 66]

# Logic: Initialize max and min with the first element
max_val = numbers[0]
min_val = numbers[0]

# Loop through the list to compare values
for num in numbers:
    if num > max_val:
        max_val = num
    if num < min_val:
        min_val = num

# Output
print(f"List: {numbers}")
print(f"Max: {max_val}")
print(f"Min: {min_val}")
```
